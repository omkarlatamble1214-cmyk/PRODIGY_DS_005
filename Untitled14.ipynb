```python
# PRODIGY_DS_05
# Traffic Accident Analysis

import pandas as pd
import matplotlib.pyplot as plt

# Sample Traffic Accident Dataset
data = {
    'Weather': ['Clear','Rain','Clear','Fog','Rain','Clear','Fog','Rain','Clear','Clear'],
    'Road_Condition': ['Dry','Wet','Dry','Wet','Wet','Dry','Wet','Wet','Dry','Dry'],
    'Time_of_Day': ['Morning','Night','Evening','Morning','Night','Afternoon','Evening','Night','Morning','Afternoon'],
    'Accidents': [12,18,10,7,20,15,8,22,14,13]
}

df = pd.DataFrame(data)

print(df)

# Weather Analysis
weather_counts = df.groupby('Weather')['Accidents'].sum()

plt.figure(figsize=(6,4))
weather_counts.plot(kind='bar')
plt.title('Accidents by Weather Condition')
plt.xlabel('Weather')
plt.ylabel('Number of Accidents')
plt.savefig('weather_analysis.png')
plt.show()

# Road Condition Analysis
road_counts = df.groupby('Road_Condition')['Accidents'].sum()

plt.figure(figsize=(6,4))
road_counts.plot(kind='bar')
plt.title('Accidents by Road Condition')
plt.xlabel('Road Condition')
plt.ylabel('Number of Accidents')
plt.savefig('road_condition_analysis.png')
plt.show()

# Time of Day Analysis
time_counts = df.groupby('Time_of_Day')['Accidents'].sum()

plt.figure(figsize=(6,4))
time_counts.plot(kind='bar')
plt.title('Accidents by Time of Day')
plt.xlabel('Time')
plt.ylabel('Number of Accidents')
plt.savefig('time_of_day_analysis.png')
plt.show()

print("Task Completed Successfully")
```
